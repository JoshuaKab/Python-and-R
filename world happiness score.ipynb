```python
#load python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load and read the dataset
df = pd.read_csv("world_happiness_2015.csv")
df.head()
df.shape
```

```python
#checking for data type per column
df.info()
```

```python
#detecte the missing values
df.isnull().sum()
```

```python
#list of all the regions

df['Region'].unique()
```

```python
#list of all the countries
df['Country'].unique()
```

```python
# Histrogram distribution
df.hist(figsize=(10,10))
plt.tight_layout()
plt.show()
```

```python
#plot the region
plt.figure(figsize=(10,8))
sns.countplot(x='Region', data=df, order=df['Region'].value_counts().index, palette='viridis')
plt.title("freqent region in the dataset")
plt.xlabel("Region")
plt.ylabel("count")
plt.xticks(rotation=45)
plt.show()
```

```python
#statistic descriptive in all the numeric column
df.describe()
```

```python
#correlation matrix
data = df.drop(['Region','Country'],axis=1)

plt.figure(figsize=(15,12))
sns.heatmap(data.corr(), annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("correlation heatmap")
plt.show()
```

```python
#scatterplot on happiness vs family column
plt.figure(figsize=(10,6))
sns.scatterplot(x='Happiness Score', y='Family', data=df, hue='Region')
plt.title('happines score vs family')
plt.xlabel('hsppyness score')
plt.ylabel('family')
plt.show()
```

```python
#scatterplot on happiness score vs GDP per capital
plt.figure(figsize=(10,6))
sns.scatterplot(x='Happiness Score', y='Economy (GDP per Capita)', data=df, hue='Region')
plt.title('happines score vs Economy (GDP per Capita)')
plt.xlabel('hsppyness score')
plt.ylabel('family')
plt.show()
```

```python
#scatterplot of happiness score vs Generosity
plt.figure(figsize=(10,6))
sns.scatterplot(x='Happiness Score', y='Generosity', data=df, hue='Region')
plt.title('happines score vs Generosity')
plt.xlabel('happyness score')
plt.ylabel('Gnerosity')
plt.show()
```

```python
#scatterplot of happiness vs GDP per capita
plt.figure(figsize=(10,6))
sns.scatterplot(x='Happiness Score', y='Economy (GDP per Capita)', data=df, hue='Region')
plt.title('happines score vs Generosity')
plt.xlabel('hsppyness score')
plt.ylabel('Gnerosity')
plt.show()
```

```python
#Pie chart distritution segment per region
region_pie = df.groupby('Region')['Happiness Score'].agg('mean')

lab = ['Australia and New Zealand ','Central and Eastern Europe ','Eastern Asia','Latin America and Caribbean '
          ,'Middle East and Northern Africa', 'North America ','Southeastern Asia ','Southern Asia','Sub-Saharan Africa'
          ,'Western Europe']

plt.pie(region_pie, labels=lab, autopct='%1.2f%%')
plt.show()
```

```python
#scatterplot GDP per capita vs happiness score per region
g = sns.FacetGrid(df, col="Region", height=4.5, aspect=.60)
g.map_dataframe(sns.scatterplot, x="Economy (GDP per Capita)", y="Happiness Score")
g.add_legend()
```

```python
#select the features column 
cols = ['Happiness Score','Standard Error', 'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
       'Generosity', 'Dystopia Residual']
```

# North_America

```python
North_america=df[df['Region']=='North America']
cols=['Happiness Score','Standard Error', 'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
       'Generosity', 'Dystopia Residual']
for col in cols:
    print(col)
    a=North_america.groupby('Country')[col].max().sort_values(ascending=False).head(2)
    print(a)
    print('-'*40)
    print('\n')
```
